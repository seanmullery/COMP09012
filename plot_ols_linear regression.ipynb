
# Linear Regression Example
The example below uses only the first feature of the `diabetes` dataset,
in order to illustrate the data points within the two-dimensional plot.
The straight line can be seen in the plot, showing how linear regression
attempts to draw a straight line that will best minimize the
residual sum of squares between the observed responses in the dataset,
and the responses predicted by the linear approximation.

The coefficients, residual sum of squares and the coefficient of
determination are also calculated.


```python
# Code source: Jaques Grobler
# License: BSD 3 clause

import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

print(f'diabetes_X shape is: {diabetes_X.shape}')

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 5]
print(f'diabetes_X shape is now: {diabetes_X.shape}')
```

```python
# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

print(diabetes_X_train.shape)
print(diabetes_X_test.shape)

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

print(diabetes_y_train.shape)
print(diabetes_y_train.max())
print(diabetes_y_test.shape)
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)
#regr.coef_[0]=20
#regr.intercept_=20
# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)
```

$$y=mx+c$$
$$y=ax_1 +bx_2 +c$$

```python
# The coefficients
print("Coefficients: \n", regr.coef_)

print(f'Y intercept: {regr.intercept_}')
print(regr.coef_.shape)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.scatter(diabetes_X_train, diabetes_y_train, color="orange")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

#plt.xticks(())
#plt.yticks(())

plt.grid()
plt.show()
```

```python
turkey_X = np.linspace(5,20, 100)
#print(f'turkey_X is: {turkey_X}')
#print(f'turkey_X shape: {turkey_X.shape}')
turkey_X = turkey_X[:, np.newaxis]
print(turkey_X.shape)
noise = np.random.normal(0, 10, 100)
print(f'max noise is :{noise.min()}')
turkey_y = 60*turkey_X[:,0]**(2.0/3.0)/1.5
print(f'y shape is {turkey_y.shape}')
turkey_y = turkey_y + noise

n = 2 # for 2 random indices
#Randomising and spliting into train test
train_index = np.random.choice(100, 80, replace=False)
print(f'train_index: {train_index}')
turkey_y_train  = turkey_y[train_index]
turkey_X_train = turkey_X[train_index]
turkey_y_test =  np.delete(turkey_y, train_index)
turkey_X_test = np.delete(turkey_X, train_index)
turkey_X_test = turkey_X_test[:, np.newaxis]
print(f'turkey_y_train :{turkey_y_train.shape}')
print(f'turkey_y_test: {turkey_y_test.shape}')
print(f'turkey_X_train :{turkey_X_train.shape}')
print(f'turkey_X_test: {turkey_X_test.shape}')
x = np.linspace(0,20,100)
y = 20*x +20
plt.scatter(turkey_X_test, turkey_y_test, color='blue', alpha=0.5)
plt.scatter(turkey_X_train, turkey_y_train, color='red', alpha= 0.5)
plt.plot(x,y)

plt.scatter
plt.show()
```

```python
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(turkey_X_train, turkey_y_train)

# Make predictions using the testing set

print("Coefficients: \n", regr.coef_)
print(regr.coef_)
print(f'Y intercept: {regr.intercept_}')
print(regr.coef_.shape)
plt.scatter(turkey_X_test, turkey_y_test, color='blue', alpha=0.5)
plt.scatter(turkey_X_train, turkey_y_train, color='red', alpha= 0.5)
plt.plot(x,y)
model = x*regr.coef_[0] + regr.intercept_

plt.plot(x, model)
plt.scatter
plt.show()
```

```python



turkey_y_pred =  regr.predict(turkey_X_test)
print("Mean squared error: %.2f" % mean_squared_error(turkey_y_test, turkey_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(turkey_y_test, turkey_y_pred))

plt.scatter(turkey_X_train, turkey_y_train, color='blue')
plt.plot(x,y, color='red',linewidth=3)
model = x*regr.coef_[0] + regr.intercept_
plt.scatter(turkey_X_test, turkey_y_test, color='yellow')
plt.plot(x, model, color='green',linewidth=3, alpha=0.5)
plt.scatter(turkey_X_test, turkey_y_pred, color=('magenta'))

plt.grid()
plt.show()
```
